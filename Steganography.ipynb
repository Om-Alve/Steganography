```python
import cv2
import matplotlib.pyplot as plt
```

```python
def encrypt(targetpath,encryptingimgpath,outpath):
    img = cv2.imread(targetpath)
    img = cv2.resize(img,(300,300))
    eimg = cv2.imread(encryptingimgpath) 
    eimg = cv2.resize(eimg,(300,300))
    highvals = img - (img%16)
    targetvals = highvals//16
    outimg = (eimg - (eimg % 16)) + targetvals
    cv2.imwrite(outpath,outimg)
```

```python
def decrypt(encryptedimg,outpath):
    eimg = cv2.imread(encryptedimg)
    decrypted = (eimg%16) * 16
    cv2.imwrite(outpath,decrypted)
```

```python
encrypt('cat.jpg','dog.jpg','out.png')
```

```python
decrypt('out.png','decrypted.png')
```

```python
img = cv2.imread('out.png')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
img = cv2.imread('decrypted.png')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```
